```python
contacts={
    'Alice':123456789
}
```

```python


def Add_Contact():
    name=input("Enter the name: ")
    while True:
        number=input("Enter your number(10 digits): ")
        if number.isdigit() and len(number)==10:
            break
        else:
            print("Invalid number. Please enter a 10 digit number.")
    contacts[name]=number

def view_details(name1):
    if name1 not in contacts:
        print(f"{name1} not found in book.")
    else:
        print("\n contact number: ")
        for name,number in contacts.items():
            if name==name1:
                print(f"{name}->{number}")
                
def search_contact(name):
    if name not in contacts:
        print(f"{name} does not exist in book.")
    else:
        print("{} contact found.".format(name))
        print("Do you want to see the details.")
        details=input("Y/N :")
        if details.lower()=='y':
            view_details(name)
        

def update_contact(name):
    if name in contacts:
        contacts[name]=input("Enter new number: ")
        print("Contact updated successfully!")
    else:
        print("Contact not found.")

def delete_contact(name):
    if name in contacts:
        del contacts[name]
        print(f"{name} deleted successfully.")
    else:
        print(f"{name} does not exist.")

def view_all_contacts():
    if not contacts:
        print("No contacts found.")
    else:
        print("\nAll contacts:")
        for name,number in contacts.items():
            print(f"{name} -> {number}")

def exit():
    print("Thank you for using Contact Book.")
        
    
```

```python
def contact_Book():
    print("---Contact Book Menu---")
    print("1. Add Contact")
    print("2. View Contact")
    print("3. Search Contact")
    print("4. Update Contact")
    print("5. Delete Contact")
    print("6. View all contacts")
    print("0.Exit")
    
    while True:
        choice=int(input("Enter your choice: "))
        if choice==1:
            Add_Contact()
        elif choice==2:
            name=input("Enter name: ")
            view_details(name)
        elif choice==3:
            name=input("Enter name: ")
            search_contact(name)
        elif choice==4:
            name=input("Enter name: ")
            update_contact(name)
        elif choice==5:
            name=input("Enter name: ")
            delete_contact(name)
        elif choice==6:
            password=input("Enter password to view all contacts:")
            if password=="admin123@":
                view_all_contacts()
            else:
                print("Incorrect password. Access denied")
        elif choice==0:
            exit()
            break
        
    
    
```
