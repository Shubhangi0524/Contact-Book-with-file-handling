```python
account={}
```

```python


def create_account():
    acc_name=input("Enter Account holder name: ").strip().lower()
    balance=float(input("Enter initial balance: "))
    account[acc_name]=balance
    print("Account created Successfully!")

def check_balance(name):
    if name.lower() in account:
        print(name)
        print("Current Balance: ",account[name])
    else:
        print("Account not found.")

def deposit(name):
    name=name.lower()
    if name in account:
        depo=float(input("Enter the amount to deposit: "))
        account[name]+=depo
        print("Deposit successful!")
    else:
        print("Account not found.")

def withdraw(name,amount):
    name=name.lower()

    if name in account:
        if account[name]<amount:
            print("Insufficient Balance!")
            print("Check balance")
            check=input("Y/N").lower()
            if check=="y":
                check_balance(name)
        else:
            account[name]-=amount
            print("Withdraw Successful")
            print("Remaining Balance: ",account[name])

def exit():
    print("Thank you for visiting the $$ bank!")
        
```

```python
def Banking_System():
    print("--- Banking System ---")
    print("1. Create Account")
    print("2. Check Balance")
    print("3. Deposit Money")
    print("4. Withdraw Amount")
    print("0. Exit")
    attempt=5
    while attempt:
        choice=int(input("Enter your choice: "))
        if choice==1:
            create_account()
            attempt-=1
        elif choice==2:
            name=input("Enter name: ")
            check_balance(name)
            attempt-=1
        elif choice==3:
            name=input("Enter name: ")
            deposit(name)
            attempt-=1
        elif choice==4:
            name=input("Enter name: ")
            amount=float(input("Enter amount to withdraw: "))
            withdraw(name,amount)
            attempt-=1
    
        else:
            attempt=0
            exit()
            break
    else:
        print("Limit exceeded, try again after 1 hour")
        
        
    
Banking_System()
```
